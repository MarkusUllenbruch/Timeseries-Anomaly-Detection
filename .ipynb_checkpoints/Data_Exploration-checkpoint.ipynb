# Data Exploration

```python
import pandas as pd
import shutil
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import MeanSquaredError, MeanAbsoluteError
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
shutil.unpack_archive('Data/ECG5000.zip', './Data')
```

```python
train = pd.read_csv('Data/ECG5000_TRAIN.txt', delimiter= '\s+', index_col=False, header=None)
test = pd.read_csv('Data/ECG5000_TEST.txt', delimiter= '\s+', index_col=False, header=None)
```

```python
# Concatenate the Datasets to one Dataset
data = pd.concat((train, test))
print('Number Samples:', data.shape[0], '\nNumber Timesteps:', data.shape[1])
```

```python
data.columns = ['target'] + [str(i+1) for i in range(140)]
```

### Data Distribution

```python
sns.countplot(data=data, x='target', palette=['green', 'blue', 'blue', 'blue', 'blue'])
plt.savefig('Plots/Data_distribution.png')
plt.show()
print(data.target.value_counts())
```

- Target 1.0 is the Healthy sample
- 2.0, 3.0, 4.0, 5.0 are the Anomalie Samples

```python
# Split Data into the targets
data_1 = data[data.target == 1]
data_2 = data[data.target == 2]
data_3 = data[data.target == 3]
data_4 = data[data.target == 4]
data_5 = data[data.target == 5]
```

### Mean and Standard Deviation per Label

```python
def stats(df, name, color='blue', std_color='lightblue'):
    """Plot Mean sample of each target and the one-standard-deviation boundaries"""
    df = np.array(df)
    mean = np.mean(df, axis=0)
    std = np.std(df, axis=0)
    plt.fill_between(np.arange(0,141), mean+std, mean-std, color=std_color)
    plt.plot(mean, color=color)
    plt.xlabel('Timestep')
    plt.ylabel('Feature Value')
    plt.savefig('Plots/' + name + '.png')
    plt.show()
```

#### Healthy Data

```python
stats(data_1, color='green', std_color='lightgreen', name='1')
```

#### Anomalous Data

```python
stats(data_2, name='2')
```

```python
stats(data_3, name='3')
```

```python
stats(data_4, name='4')
```

```python
stats(data_5, name='5')
```

### Mean and Standard Deviation per Anomaly and Normal Data

```python
def stats2(df):
    """Plot Mean sample of each target and the one-standard-deviation boundaries"""
    df_normal = np.array(df[df.target==1])
    df_anomaly = np.array(df[df.target!=1])
    
    mean_normal = np.mean(df_normal, axis=0)
    std_normal = np.std(df_normal, axis=0)
    mean_anormal = np.mean(df_anomaly, axis=0)
    std_anormal = np.std(df_anomaly, axis=0)
    
    plt.fill_between(np.arange(0,141), mean_normal+std_normal, mean_normal-std_normal, color='lightgreen')
    plt.fill_between(np.arange(0,141), mean_anormal+std_anormal, mean_anormal-std_anormal, color='lightblue')
    
    plt.plot(mean_anormal, color='darkblue', label='Anomaly')
    plt.plot(mean_normal, color='green', label='Healthy')
    
    plt.xlabel('Timestep')
    plt.ylabel('Feature Value')
    plt.legend()
    plt.savefig('Plots/Dataset.png')
    plt.show()
    
stats2(data)
```

Looks like Autoencoder and Variational Autoencoder Networks will be able to differ from anomalies and normal data, since we are able to see the differences with our eyes. We will see in the process, whether this hypothesiss is true. Let's find out in another notebook!
